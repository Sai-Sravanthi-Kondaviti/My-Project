```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Load the dataset
file_path = "Placement_Data.csv"  # Ensure the file is uploaded in Colab
df = pd.read_csv(file_path)

# Creating interactive widgets
plot_type = widgets.Dropdown(
    options=['Histogram', 'Scatter Plot', 'Bar Chart', 'Box Plot'],
    description='Plot Type:',
    style={'description_width': 'initial'}
)

x_axis = widgets.Dropdown(
    options=['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p'],
    description='X-axis:',
    style={'description_width': 'initial'}
)

y_axis = widgets.Dropdown(
    options=['salary', 'ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p'],
    description='Y-axis:',
    style={'description_width': 'initial'}
)

def plot_graph(plot_type, x_axis, y_axis):
    plt.figure(figsize=(8, 5))

    if plot_type == "Histogram":
        plt.hist(df[x_axis], bins=10, color='blue', edgecolor='black')
        plt.xlabel(x_axis)
        plt.ylabel("Frequency")
        plt.title(f"Histogram of {x_axis}")

    elif plot_type == "Scatter Plot":
        plt.scatter(df[x_axis], df[y_axis], color='green', alpha=0.5)
        plt.xlabel(x_axis)
        plt.ylabel(y_axis)
        plt.title(f"Scatter Plot: {x_axis} vs {y_axis}")

    elif plot_type == "Bar Chart":
        df_grouped = df.groupby(x_axis)[y_axis].mean()
        df_grouped.plot(kind='bar', color='purple')
        plt.xlabel(x_axis)
        plt.ylabel(y_axis)
        plt.title(f"Bar Chart of {y_axis} by {x_axis}")

    elif plot_type == "Box Plot":
        plt.boxplot(df[x_axis])
        plt.ylabel(x_axis)
        plt.title(f"Box Plot of {x_axis}")

    plt.grid()
    plt.show()

# Create interactive GUI
widgets.interactive(plot_graph, plot_type=plot_type, x_axis=x_axis, y_axis=y_axis)
```

```python
# Install required libraries
!pip install numpy pandas matplotlib seaborn

# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # For better visualization

# Load the dataset (Ensure file is uploaded in Colab)
file_path = "Placement_Data.csv"
df = pd.read_csv(file_path)

# Display the first 5 rows of the dataset
print("ðŸ”¹ First 5 Rows of the Dataset:")
display(df.head())

# Basic dataset information
print("\nðŸ”¹ Dataset Information:")
df.info()

# Check for missing values
print("\nðŸ”¹ Missing Values in the Dataset:")
print(df.isnull().sum())

# Fill missing values (if any)
df.fillna(0, inplace=True)  # Replace NaN with 0

# Display basic statistical information
print("\nðŸ”¹ Basic Statistical Summary:")
display(df.describe())

# Count of students placed vs not placed
plt.figure(figsize=(6,4))
sns.countplot(x=df['status'], palette='coolwarm')
plt.title("Placement Status Count")
plt.xlabel("Placement Status")
plt.ylabel("Count")
plt.show()

# Distribution of salaries (for placed students only)
plt.figure(figsize=(7,5))
sns.histplot(df[df['status']=='Placed']['salary'], bins=10, kde=True, color="green")
plt.title("Salary Distribution for Placed Students")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()

# Scatter plot of Degree Percentage vs Salary (Only for placed students)
plt.figure(figsize=(7,5))
df_filtered = df[df['status']=='Placed']  # Consider only placed students
plt.scatter(df_filtered['degree_p'], df_filtered['salary'], alpha=0.5, color='blue')
plt.title("Scatter Plot: Degree Percentage vs Salary")
plt.xlabel("Degree Percentage")
plt.ylabel("Salary")
plt.grid(True)
plt.show()

# âœ… FIX: Select only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=[np.number])

# Correlation matrix heatmap (excluding categorical columns)
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
